# Representación Gráfica de datos

### Si a la función de trazado se le da una matriz de datos, la usará como coordenadas en el eje vertical, y utilizará el índice de cada punto de datos en el array como la coordenada horizontal.

```python
plt.plot([1, 2, 5, 7, 8, 3, 1])
plt.show()
```

```python
lista1= [1, 2, 3, 4, 5, 6, 7]
lista2= [5, 4, 5, 7, 8, 3, 1]
#tambien se puede proporcionar una lista de valores para el eje x y el eje y
data = plt.plot(lista1, lista2)
plt.show(data)
#precaucion: las listas deben tener el mismo numero de elementos
```

```python
#se puede ajustar los valores de los ejes para que se muestren los valores que la figura no se vea tan ajustada
plt.plot([1, 2, 3, 4, 5, 6, 7], [2, 2, 5, 7, 8, 3, 1])
plt.axis([0, 8, 0, 8])
plt.show()
```

```python
#podemos ponerle un titulo a la grafica y mostrar una cuadricula
plt.plot([1, 2, 3, 4, 5, 6, 7], [2, 2, 5, 7, 8, 3, 1])
plt.axis([0, 10, 0, 10])
plt.title("Grafica de ejemplo")
plt.grid(True)
plt.show()
```

```python
#tambien podemos representar funciones matematicas
import numpy as np
x = np.linspace(2,-2, 500)
y = x**2
plt.plot(x, y)
plt.title("Funcion cuadratica")
plt.grid(True)
plt.xlabel("Eje x")
plt.ylabel("Eje y (x^2)")
plt.show()
```

```python
# podemos representar varias funciones en una misma grafica y ponerle un color diferente a cada una
import numpy as np
x = np.linspace(2,-2, 500)
y = np.sin(x)
y2 = x+1
plt.plot(x, y,'b--', x, y2, 'g')
plt.show()
```

```python
#otra forma de representar varias funciones en una misma grafica es "acumularlas" en la misma grafica
import numpy as np
x = np.linspace(2,-2, 500)
y = np.sin(x)
y2 = x+1
plt.plot(x, y,'b--')
plt.plot(x,y2, 'g')
#si nos fijamos, la grafica es la misma que la anterior, solo que en esta ocasion se han hecho dos llamadas a la funcion plot 
#tambien podemos ponerle una leyenda a la grafica
plt.legend(["sin(x)", "x+1"])
#gracias a estos podemos identificar que funcion es cada una
plt.show()
```

```python
#Tambien podemos separar las graficas en subgraficas
import numpy as np
x = np.linspace(2,-2, 500)
y = np.sin(x)
y2 = x+1
plt.subplot(1, 2, 1)
plt.plot(x, y,'b--')
plt.title("Funcion seno")
plt.subplot(1, 2, 2)
plt.plot(x,y2, 'g')
plt.title("Funcion lineal")
plt.show()
```

```python
#Podemos editar para que las graficas sean mas grandes
import numpy as np
plt.figure(figsize=(20,10))
x = np.linspace(2,-2, 500)
y = np.sin(x)
y2 = x+1
plt.subplot(1, 2, 1)
plt.plot(x, y,'b--')
plt.xlabel("Eje x1", fontsize=15)
plt.ylabel("Eje y1", fontsize=15)
plt.title("Funcion seno", fontsize=15)

plt.subplot(1, 2, 2)
plt.plot(x,y2, 'g')
plt.xlabel("Eje x2", fontsize=15)
plt.ylabel("Eje y2", fontsize=15)
plt.title("Funcion lineal", fontsize=15)
plt.show()
```
