# Representación Gráfica de datos

### Si a la función de trazado se le da una matriz de datos, la usará como coordenadas en el eje vertical, y utilizará el índice de cada punto de datos en el array como la coordenada horizontal.

```python
plt.plot([1, 2, 5, 7, 8, 3, 1])
plt.show()
```

```python
#tambien se puede proporcionar una lista de valores para el eje x y el eje y
data = plt.plot([1, 2, 3, 4, 5, 6, 7], [2, 2, 5, 7, 8, 3, 1])
plt.show(data)
#precaucion: las listas deben tener el mismo numero de elementos
```

```python
#se puede ajustar los valores de los ejes para que se muestren los valores que la figura no se vea tan ajustada
plt.plot([1, 2, 3, 4, 5, 6, 7], [2, 2, 5, 7, 8, 3, 1])
plt.axis([0, 10, 0, 10])
plt.show()
```

```python
#podemos ponerle un titulo a la grafica y mostrar una cuadricula
plt.plot([1, 2, 3, 4, 5, 6, 7], [2, 2, 5, 7, 8, 3, 1])
plt.axis([0, 10, 0, 10])
plt.title("Grafica de ejemplo")
plt.grid(True)
plt.show()
```

```python
#tambien podemos representar funciones matematicas
import numpy as np
x = np.linspace(2,-2, 500)
y = x**2
plt.plot(x, y)
plt.title("Funcion cuadratica")
plt.grid(True)
plt.xlabel("Eje x")
plt.ylabel("Eje y (x^2)")
plt.show()
```
